```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../csv/delays_reduced.csv')
df.head()
```

```python
df["delay_rate"] = np.where(
    df["arr_flights"] > 0,
    df["arr_del15"] / df["arr_flights"],
    np.nan
)
df["delay_rate"] = df["delay_rate"] * 100
```

```python
df["avg_delay_min"] = df["arr_delay"] / df["arr_flights"]
```

```python
season_map = {
    12: "Winter", 1: "Winter", 2: "Winter",
    3: "Spring", 4: "Spring", 5: "Spring",
    6: "Summer", 7: "Summer", 8: "Summer",
    9: "Fall", 10: "Fall", 11: "Fall"
}

df["season"] = df["month"].map(season_map)
```

```python
df.to_csv('../csv/delays_updated.csv', index=False, encoding='utf-8')
```
